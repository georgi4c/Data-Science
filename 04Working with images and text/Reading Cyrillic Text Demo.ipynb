# Reading Cyrillic Text
## Demo: 

```python
with open("data/beliyat_gushter.txt", "r", encoding = "utf-8") as file:
    text = file.read()
```

```python
print(text[:2000])
```

```python
words = re.split("\W+", text)
words = [w for w in words if w != ""]
words = [w.lower() for w in words]
```

```python
Counter(words).most_common(100)
```
